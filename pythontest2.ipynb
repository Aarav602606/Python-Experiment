```python
## Q1 Class variable vs instance variable ?
```

```python
class student:
    school_name = "UPES"

    def __init__ (self, name, age):
        self.name = name
        self.age = age

        s1 = student("Aman", 20)
        s2 = student("Riya", 21)

        print("school:", s1.school_name)
        print("school:", s2.school_name)

        print("student 1:", s1.name, s1.age)
        print("student 2:", s2.name, s2.age)
```

```python
## Q3 Static Method ?
```

```python
class mathhelper:
    @staticmethod
    def is_even(num):
        if num % 2 == 0:
            return True
        else:
            return False
print(mathhelper.is_even(8))
print(mathhelper.is_even(11))
```
