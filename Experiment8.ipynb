```python
with open("name.txt", "w") as f:
    n = int(input("Enter number of names: "))
    for i in range(n):
        name = input()
        f.write(name + "\n")
```

```python
with open("name.txt", "r") as f:
    names = f.readlines()

print("Total names:", len(names))
```

```python
## Question 1c:
```

```python
count = 0

with open("name.txt", "r") as f:
    for line in f:
        if line.strip()[0].lower() in "aeiou":
            count += 1

print("Names starting with vowel:", count)
```

```python
## Question 1d:
```

```python
longest = ""

with open("name.txt", "r") as f:
    for line in f:
        name = line.strip()
        if len(name) > len(longest):
            longest = name

print("Longest name:", longest)
```

```python
with open("numbers.txt", "w") as f:
    n = int(input("Enter count of numbers: "))
    for i in range(n):
        num = input()
        f.write(num + "\n")
```

```python
## Question 2b:
```

```python
with open("numbers.txt", "r") as f:
    nums = [int(x.strip()) for x in f]

print("Maximum:", max(nums))
```

```python
## Question 2c:
```

```python
count = 0
with open("numbers.txt", "r") as f:
    numbers = [int(x.strip()) for x in f]

for num in numbers:
    if num > 100:
        count += 1

print("Numbers greater than 100:", count)
```

```python
## Question 3c:
```

```python
total_area = 0

with open("city.txt", "r") as f:
    for line in f:
        data = line.split()
        total_area += float(data[2])

print("Total area:", total_area)
```

```python
## Question 4:
```

```python
n = int(input())

for i in range(n):
    try:
        a, b = input().split()
        result = int(a) // int(b)
        print(result)
    except ZeroDivisionError as e:
        print("Error Code:", e)
    except ValueError as e:
        print("Error Code:", e)
```

```python
try:
    filename = input("Enter file name: ")
    f = open(filename, "r")
    data = f.read()
    print(data)
    f.close()
except FileNotFoundError:
    print("File not found")
except PermissionError:
    print("Permission denied")
except IOError:
    print("Input Output error occurred")
except Exception as e:
    print("Error:", e)
```
