```python
##Question 1:

def find_max_min(numbers):
    maximum = numbers[0]
    minimum = numbers[0]
    
    for num in numbers:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num
    
    return maximum, minimum

nums = list(map(int, input("Enter numbers: ").split()))
result = find_max_min(nums)
print("Maximum:", result[0])
print("Minimum:", result[1])
```

```python
##Question 2:

def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total += i ** 3
    return total

n = int(input("Enter number: "))
print("Sum of cubes:", sum_of_cubes(n))
```

```python
##Question 3:

def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)

n = int(input("Enter number: "))
print_numbers(n)
```

```python
##Question 4:

def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

n = int(input("Enter number of terms: "))

for i in range(n):
    print(fibonacci(i), end=" ")
```

```python
##Question 5:

volume = lambda r, h: (1/3) * 3.14 * r * r * h

r = float(input("Enter radius: "))
h = float(input("Enter height: "))

print("Volume of cone:", volume(r, h))
```

```python
##Question 6:

nums = list(map(int, input("Enter numbers: ").split()))

result = (lambda lst: (max(lst), min(lst)))(nums)

print(result)
```

```python
##Question 7c:

def total_sum(*numbers):
    total = 0
    for num in numbers:
        total += num
    print("Sum:", total)

total_sum(10, 20, 30)
```
